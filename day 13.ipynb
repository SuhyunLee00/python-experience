```python
import random

#total money
M= 1000000

#beginning price of the stock
S= 10000

#total stock that I have
TS=0

hour = 9
minute = 30

while hour < 16:
    E=random.uniform(1.1,0.9)
    S=round(S*E)
    
    print(hour," : " ,minute)
    print(f'현재가격{S}')

    choice = input('buy or sell or skip ')

    if choice == 'buy':
        count = int(input('how many? '))
        # 사려고 하는게 총 얼만지 계산
        Totalprice=S*count
        # 그 가격이 지갑에 있는지 확인
        # 있다면 구매를 하고 없으면 다시 선택하게
        # 구매를 했다면 보유한 주식량이 변화
        if Totalprice<=M:
            M=M-Totalprice
            TS=TS+count
        else:
            continue
        
        
    elif choice == 'sell':
        count = int(input('how many? '))
        #파려고한 총갯수가 지갑에 있는지
        #있다면 몇 주를 팔건지
        #없으면 패스
        #팔고난 후의 주식 보유량
        #지갑의 변화
        #시세차이익*
        if TS>=count:
            TS=TS-count
            M=M+S*count
        else: 
            continue
            

    elif choice == 'skip':
        pass
    else:
        
        pass

    minute+=30
    if minute == 60:
        hour+=1
        minute=0

    print(f'총자산:{M}')
    print(f'보유주식량:{TS}')
    print()

        
```

```python
import random
class Stock:
    def __init__(self, S, E1, E2, name):
        self.S=S
        self.E1=E1
        self.E2=E2
        self.TS=0
        self.name=name
    
    def select_cost(self):
        self.E=random.uniform(self.E1, self.E2)
        self.S=round(self.S * self.E)
    def buy_action(self, M):
        while True:
            try:
                count = int(input('how many? '))
                Totalprice = self.S * count
                if Totalprice <= M:
                    M = M - Totalprice
                    self.TS = self.TS + count
                    return M
                else:
                    continue

            except:
                print('올바른 형태로 입력하시오.')
    def sell_action(self, M):
        while True:
            try:
                count = int(input('how many? '))
                if self.TS>=count:
                    self.TS=self.TS-count
                    M=M+self.S*count
                    return M
                else: 
                    continue
            except:
                print('올바른 형태로 입력하시오.')
    def end_action(self):
        return self.TS*self.S
    
    
```

```python
import random

#total money
M= 1000000

stock_list = []

#beginning price of the stock
stock_list.append(Stock(60000, 1.1, 0.8, 'samsung'))
stock_list.append(Stock(180000, 1.2, 0.8, 'apple'))

#total stock that I have

hour = 9
minute = 30

while hour < 16:
    
    print(hour," : " ,minute)
    for stock in stock_list:
        stock.select_cost()
        print(f'{stock.name} : 현재가격{stock.S}')

    for stock in stock_list:
        while True:
            choice = input(f'{stock.name} : buy or sell or skip ')
            if choice in ['buy', 'sell','skip']:
                break

        if choice == 'buy':
            M = stock.buy_action(M)

        elif choice == 'sell':
            M = stock.sell_action(M)

    minute+=30
    if minute == 60:
        hour+=1
        minute = 0
    
    print(f'현금 : {M}')
    for stock in stock_list:
        print(f'{stock.name} 보유주식량 : {stock.TS}')
        
    print()

MM = M
for stock in stock_list:
    MM += stock.end_action()
print(MM)
```
